```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_excel('新数据.xlsx')
```

```python
df
```

**透视图**

```python
pd.pivot_table(data=df,index='学名',columns='种加词',values='科拉丁名',
               fill_value=0,aggfunc=np.sum,margins=True,margins_name='总计求和值')
```

**交叉表**

```python
pd.crosstab(index=df['种加词'],columns=df['科中文名'],margins=True)
```
