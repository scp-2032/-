```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,100,1000)
```

```python
y=np.sin(x)
```

```python
plt.plot(x)
```

```python
plt.plot(x,y,c='red',lw=2,ls='-')#lw是线条粗度
plt.show()
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
```

```python
plt.plot(x,y,ls=':',lw=2,label='x和y的关系')
plt.legend
```

解决matplotlib中文显示

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.formatter.useoffset']=False
plt.plot(x,y,ls=':',lw=2,label='x和y的关系')
plt.legend(loc='upper center')
plt.show()
```

添加标签和图例

```python
x=np.arange(0,1.1,0.1)
```

```python
y=x**2
```

```python
plt.figure(figsize=(6.4,4.84),dpi=100
           ,facecolor='white')#dpi是分辨率，facecolor是边框
plt.title('这是一幅画')#标题
plt.xlabel('x1')#轴的标签
plt.ylabel('y')
plt.xlim([0,1])#x轴的范围
plt.ylim([0,1])
plt.xticks(np.arange(0,1.2,0.2))#设置刻度
plt.yticks(np.arange(0,1.2,0.2))
plt.plot(x,y,label='y=x*2')
plt.legend(loc='best')#添加图例，并且自动放置在合适的位置
#plt.savefig('位置')
plt.show()
```
