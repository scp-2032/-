```python
from matplotlib import pyplot as plt
#设置中文显示
from matplotlib.font_manager import FontProperties
my_font=FontProperties(fname='/usr/share/fonts/opentype/noto/NotoSerifCJK-Bold.ttc')
x=range(2,26,2)

y=[15,13,14,5,17,28,26,26,24,22,18,15]



#设置图片大小
fig=plt.figure(figsize=(20,8),dpi=80)
#绘图
plt.plot(x,y)
#设置刻度,两种方法
_xtick_labels=['{}点'.format(i) for i in x]
plt.xticks(x,_xtick_labels,fontproperties=my_font)

plt.yticks(range(min(y),max(y)+1))
#保存图片
#plt.savefig(".sig_size.png")


#添加描述信息
plt.xlabel("时间",fontproperties=my_font)
plt.ylabel("温度",fontproperties=my_font)
plt.title('10点到12点温度随时间的变化情况',fontproperties=my_font)

#增加网格线
plt.grid(alpha=0.6)
#展示结果
plt.show()
```
