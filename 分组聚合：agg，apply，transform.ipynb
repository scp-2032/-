```python
import pandas as pd
import os
import numpy as np
```

```python
val=pd.read_excel('新数据.xlsx')
```

```python
val
```

```python
val.fillna(1)
```

```python
names=['属拉丁名','属中文名','中名','学名','种加词']
```

```python
grouped=val[names].groupby('种加词')
```

**对指定数据进行相应运算**

```python
grouped.agg([np.mean,np.sum])
```

**对不同的指定行进行指定运算**

```python
grouped.agg({'学名':[np.sum,np.median],'中名':[np.mean,np.sum]})
```

**生成函数进行运算**

```python
#自定义函数
def DoubleSUM(data):
    s=data.sum()*2
    return s
#求和再乘2
```

```python
grouped.sum()['中名']
```

```python
grouped.agg({'中名':DoubleSUM})
```

**apply函数，可以指定行列**

```python
val[names].apply(np.sum,axis=0)#沿列进行求和
```

还可以使用lambda函数

```python
val[names].apply(lambda x:x[2]-x[3],axis=1)
```

```python
#transform函数
grouped.mean().transform(lambda x:x*2)
```

```python
val[names].transform(np.sum)#会报错，因为transform无法进行聚合操作
```

```python
agg：可以对分组对象进行相同的运算，也可以选择不同的聚合方法
apply；可以进行聚合计算，还可以按行计算
transform：返回与数据同样长度的行，无法进行聚合
```
